````python
sat_filepath = '/Users/alexanderhalcomb/Desktop/DSI-SF-2-alexhalcomb/datasets/state_sat_scores/sat_scores.csv'

# pure Python implementation

with open(sat_filepath,'r') as f:
    reader = csv.reader(f)
    scores = [row for row in reader]

scores_dict = {}
for index,column in enumerate(scores[0]):
    temp_list = []
    for item in scores[1:]:
        temp_list.append(item[index])
    scores_dict.update({column:temp_list})
scores_dict
````

### 5.1 Find the list of states that have `Verbal` scores greater than the average of `Verbal` scores across states

How many states are above the mean? What does this tell you about the distribution of `Verbal` scores?




````python
mean_verbal = np.mean(sat_scores_new2['Verbal'])
verbal_over_mean = sat_scores_new2[sat_scores_new2['Verbal'] > mean_verbal]
print len(verbal_over_mean)
print mean_verbal

# 24 states are above the mean of 533, which is less than half of all states. 
# This implies that the distribution is slightly positively skewed. 
````

### 5.2 Find the list of states that have `Verbal` scores greater than the median of `Verbal` scores across states
